```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('covid_toy.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df.drop('has_covid',axis=1),df['has_covid'],test_size=0.2,random_state=42)

X_train.shape
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer()


X_train_new=si.fit_transform(X_train[['fever']])
X_test_new=si.transform(X_test[['fever']])
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Mild','Strong']])

X_train_oe=oe.fit_transform(X_train[['cough']])
X_test_oe=oe.transform(X_test[['cough']])
```

```python
df['cough'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)

X_train_ohe=ohe.fit_transform(X_train[['gender','city']])
X_test_ohe=ohe.transform(X_test[['gender','city']])
```

```python
X_train_ohe.shape
```

```python
X_train_age=X_train.drop(columns=['gender','fever','cough','city']).values
X_test_age=X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age
```

```python
X_train_trf=np.concatenate((X_train_new,X_train_age,X_train_oe,X_train_ohe),axis=1)

X_test_tef=np.concatenate((X_test_new,X_test_age,X_test_oe,X_test_ohe),axis=1)
```
