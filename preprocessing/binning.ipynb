```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_predict
```

```python
df=pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
df.info()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.head()
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
dt=DecisionTreeClassifier()

dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)

accuracy_score(y_test,y_pred)
```

```python
np.mean(cross_val_predict(dt,x,y,cv=10,scoring='accuracy'))
```

```python
kbin_age=KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='kmeans')
kbin_fare=KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='kmeans')
```

```python
trf=ColumnTransformer([
    ('first',kbin_age,[0]),
    ('second',kbin_fare,[1])
])
```

```python
X_train_trf=trf.fit_transform(X_train)
X_test_trf=trf.fit(X_test)
```

```python
trf.named_transformers_['first'].n_bins_
```

```python
output=pd.DataFrame({
    'age':X_train['Age'],
    'age_trf':X_train_trf[:,0],
    'fare':X_train['Fare'],
    'fare_trf':X_train_trf[:,1]
})
```

```python
dt=DecisionTreeClassifier()

dt.fit(X_train_trf,y_train)
y_pred2=dt.predict(X_test_trf)
accuracy_score(y_test,y_pred2)
```
