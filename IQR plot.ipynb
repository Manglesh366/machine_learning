```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('placement.csv')
df.info()
df.isnull().sum()
```

```python
df.sample(10)
```

```python
def plot(col1,col2):
    plt.figure(figsize=(16,5))
    sns.displot(col1,kde=True)
    sns.displot(col2,kde=True)
    plt.show()
```

```python
plot(df['cgpa'],df['placement_exam_marks'])
```

```python
df['placement_exam_marks'].describe()
```

```python
def boxplt(col):
     sns.boxplot(col)
     plt.figsize=(12,5)
     plt.show()
```

```python
boxplt(df['cgpa'])
```

```python
percent25=df['placement_exam_marks'].quantile(0.25)
percent75=df['placement_exam_marks'].quantile(0.75)
```

```python
print(percent25
,percent75)
```

```python
iqr=percent75-percent25
print(iqr)
```

```python
lowerlimit=percent25-1.5*iqr
upperlimit=percent75+1.5*iqr
print(lowerlimit,upperlimit)

```

```python
df[df['placement_exam_marks']>upperlimit]
```

```python
new_df=df[df['placement_exam_marks']<upperlimit]
```

```python
plot(new_df['cgpa'],new_df['placement_exam_marks'])
```

```python
print(df['placement_exam_marks'].skew(),
new_df['placement_exam_marks'].skew())
```
